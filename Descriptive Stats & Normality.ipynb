```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

<strong>#1 BINOMIAL</strong>

```python
binomial_var = np.random.binomial(20, 0.5, 100)
plt.hist(binomial_var)
plt.show()
```

```python
plt.hist(binomial_var, bins=10, color='lightblue')

# Add a vertical line at the mean.
plt.axvline(binomial_var.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(binomial_var.mean() + binomial_var.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(binomial_var.mean()-binomial_var.std(), color='b', linestyle='dashed', linewidth=2) 


plt.show()
```

<strong>Lastly...</strong> 
 
 1. Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, 
and the other with a mean of 10 and standard deviation of 1.  
 2. Add them together to create a third variable.
 3. Graph the third variable using a histogram.
 4. Compute the mean and standard deviation and plot them as vertical lines on the histogram.
 5. Evaluate the descriptive statistics against the data

```python
var1 = np.random.normal(5,.5, 100)
var2 = np.random.normal(10,1, 100)
var3 = var1+var2

mean = np.mean(var3)
std = np.std(var3)

plt.hist(var3, color='maroon')
plt.axvline(x=mean,color='black', linestyle='solid', linewidth=2)
plt.axvline(x=mean+std, color='black',linestyle='dashed', linewidth=2)
plt.axvline(x=mean-std, color='black',linestyle='dashed', linewidth=2)

plt.show()
```
